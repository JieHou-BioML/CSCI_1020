```python
#Task 4.1.1
#Criterion - This parameter tests the quality of the split type chosen by the user the default split being gini. This will help the user determine how useful the data will be in making predictions due to the quality of split.
#Max_depth - The will be the max depth that the tree will be allowed to explore. If none is given, the tree will fill itself out naturally. Allowing the tree to find the right number may be the most accurate because users who given to far a depth will lead to overfitting and not giving it enough depth will lead to underfitting.
#Min_samples_split - The minimum number of samples needed to split a node within the tree. The default is 2 which is typical as a higher or lower number may skew results when dealing with lots of data.
#Min_samples_leaf - The minimum amount of samples needed to a leaf. If the number is greater than 1, this will cause issues with large amounts of data because the points will not all be individuals if they must share a node with another data point.
#Max_features - How many features will be looked at when creating a split. This will allow the user to decide if certain features are not important and therefore should not be included when making predictions which can allow for more accurate predictions depending on the quality of the data.
```

```python
#Task 4.1.2
#Fit - This will design a decision tree based off the training data. 
#Predict - This will predict a class or the regression for a single feature based off the data given to it.
#Predict_proba - The probabilty for the prediction of a class will be given based on the data given to it.
#Score - This will be average accuracy for the test data given to the model and the labels.
```

```python
from sklearn.datasets import fetch_openml
from joblib import dump, load

fmnist = fetch_openml('Fashion-MNIST', version=1)

X_train = fmnist.data[:50000]
y_train = fmnist.target[:50000]
X_valid = fmnist.data[50000:]
y_valid = fmnist.target[50000:]

# Train a decision tree classifier on the training data
tree = DecisionTreeClassifier(random_state=42, max_depth=10)
tree.fit(X_train, y_train)

# Make predictions on the training and validation data
y_train_tree_pred = tree.predict(X_train)
y_valid_tree_pred = tree.predict(X_valid)

print("Training prediction:", y_train_tree_pred)
print("Validation prediction:", y_valid_tree_pred)

# Compute the accuracy of the predictions
train_acc_tree = accuracy_score(y_train, y_train_tree_pred)
valid_acc_tree = accuracy_score(y_valid, y_valid_tree_pred)
prec_tree = precision_score(y_valid, y_valid_tree_pred, average = 'micro')
recall_tree = recall_score(y_valid, y_valid_tree_pred, average = 'micro')
f1_tree = f1_score(y_valid, y_valid_tree_pred, average = 'micro')

print("Training accuracy:", train_acc_tree)
print("Validation accuracy:", valid_acc_tree)
print("Precision Accuracy:", prec_tree)
print("Recall Score:", recall_tree)
print("f1 Score:", f1_tree)

# Predict probabilities on the training and validation data
y_train_proba = tree.predict_proba(X_train)
y_valid_proba = tree.predict_proba(X_valid)

print("Training Proba:", y_train_proba)
print("Validation Proba:", y_valid_proba)

dump(tree, 'decision_tree_model.joblib')
```

```python
#Difference between 'predict()' and 'predict_proba()'
#Predict returns a single value which is the models prediction for which class the data point will fall into while predict_proba() gives a ranges of values for the probability that the data point falls into each class.
```

```python
import gradio as gr
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Load the Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Flatten the images into a single row
x_train = x_train.reshape(x_train.shape[0], -1)
x_test = x_test.reshape(x_test.shape[0], -1)

# Define the available classifiers
classifiers = {
    "Decision Tree": DecisionTreeClassifier(random_state=42, max_depth=20, max_features = 10),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=50, learning_rate=1.0),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=2, learning_rate = 1.0),
    "KNN": KNeighborsClassifier(n_neighbors=5)
}

# Define a function to preprocess the image
def preprocess(image):
    # Resize the image to 28x28 and normalize the pixel values
    image = tf.image.resize(image, (28, 28)) / 255.0
    # Flatten the image into a single row
    image = tf.reshape(image, (1, -1))
    return image.numpy()

# Define a function to make predictions
def predict(image, classifier):
    # Preprocess the image
    image = preprocess(image)
    # Train the selected classifier on the training data
    clf = classifiers[classifier]
    clf.fit(x_train, y_train)
    # Predict the probabilities for the image using the trained classifier
    probs = clf.predict_proba(image)
    # Convert the probabilities to a list
    probs = probs.tolist()[0]
    # Return the probabilities
    return probs

# Define the input and output interfaces
inputs = [
    gr.inputs.Image(label="Image"),
    gr.inputs.Dropdown(choices=list(classifiers.keys()), label="Method")
]
output = gr.outputs.Label(num_top_classes=10)

# Create the Gradio interface
interface = gr.Interface(fn=predict, inputs=inputs, outputs=output, live=True, capture_session=True)

# Launch the interface
interface.launch(debug=True)
```
